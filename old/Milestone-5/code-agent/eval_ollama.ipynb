```python
import ollama
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel
import torch
import os
from datasets import load_dataset
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

tokeniser=AutoTokenizer.from_pretrained("meta-llama/Meta-Llama-3.1-8B-Instruct")
model= AutoModelForCausalLM.from_pretrained(
    "meta-llama/Meta-Llama-3.1-8B-Instruct",
    quantization_config=bnb_config,
    device_map="auto",
)
lora_path = "C:\\Users\\user\\Desktop\\Raghul\\DSAI\\Project\\outputs\\unsloth_lora_lama3.18b_resoningdataset"
model = PeftModel.from_pretrained(model, lora_path)
```

```python
dataset_dict=load_dataset("nvidia/OpenCodeReasoning", "split_0")
```

```python
prompt="""

You are a code-evaluation assistant. Your task is to evaluate code generated by a fine-tuned language model.

You will be given:

A programming question or task, and

The model-generated answer.

Your job is to review the provided answer and score it according to the following rubric:

Evaluation Criteria (Total: 5 points)

Correctness (0-2 points)

Does the code solve the given problem?

Does it run without errors?

Are edge cases handled appropriately?

Clarity of Reasoning (0-1 point)

Does the explanation (if provided) clearly describe the approach?

Is the reasoning easy to follow?

Step-by-Step Logic (0-1 point)

Is the solution broken down into logical steps?

Does the explanation reflect a coherent thought process?

Readability (0-1 point)

Is the code organized and well-formatted?

Are variable names, comments, and structure clear?

Output Format

Score: Your rating

Make sure you follow the output format. DON'T GIVE ANY OUTPUT OTHER THAT THAT. DON'T GIVE ANY REASONING.
"""
```

```python
output_list=[]
for i in range(1000,1100):
    code_input=dataset_dict['split_0'][i]['input']

    token_input=tokeniser(code_input,return_tensors='pt').to('cuda')
    output=model.generate(**token_input,max_new_tokens=2000)
    finetuned_code=tokeniser.decode(output[0])
    output_to_be_checked=prompt+"*********Model output begins from here***************"+finetuned_code
    response=ollama.generate(model='gpt-oss:20b',prompt=output_to_be_checked)
    output_list.append(response.response)
```
